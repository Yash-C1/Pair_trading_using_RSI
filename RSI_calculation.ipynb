```python
import pandas as pd
import numpy as np
```

```python
pairs = pd.read_csv('/content/final_pairs.csv')
stocks = pd.read_csv('/content/final_stocks.csv')
```

```python
pairs.head()
```

```python
stocks.head()
```

```python
company_1,company_2=pairs.iloc[0,1:3]
```

```python
company_1_prices = stocks[company_1]
company_2_prices = stocks[company_2]
dates = stocks['date']
```

```python
new_df = pd.DataFrame({'dates':dates,'company A':company_1_prices,'company B':company_2_prices})
```

```python
new_df.head()
```

```python
new_df.columns
```

```python
new_df
```

```python
ratio = new_df['company A']/new_df['company B']
```

```python
inverse_ratio = 1/ratio
```

```python
ratio
```

```python
new_df.insert(3,"Ratio",ratio)
new_df.head()
```

```python
# new_df.insert(4,"Inverse_Ratio",inverse_ratio)
# new_df.head()
```

```python
#change = new_df['Ratio'][i]
```

```python
new_df.insert(4,"change",[0 for i in range(len(new_df['Ratio']))])
new_df.head()
```

```python
# new_df.insert(0,"sr_no",[i for i in range(len(new_df['Ratio']))])
# new_df.head()
```

```python
new_df['Ratio'][162]
```

```python
ratio_values = new_df['Ratio']
```

```python
change = []
```

```python
for i in range(len(ratio_values)-1):
  change.append(ratio_values[i]-ratio_values[i+1])
change.append(new_df['Ratio'][162])
```

```python
len(change)
```

```python
new_df['change']=change
```

```python
new_df
```

```python
gain = []
for i in range(len(change)):
  if(change[i]>0):
    gain.append(change[i])
  else:
    gain.append(0)
```

```python
new_df['gain'] = gain
new_df
```

```python
loss = []
for i in range(len(change)):
  if(change[i]<0):
    loss.append(-1*change[i])
  else:
    loss.append(0)
```

```python
new_df['loss'] = loss
new_df
```

```python
avg_gain = [0 for i in range(14)]
```

```python
# sum(gain[len(gain)-15:len(gain)-1])
avg_gain.append(sum(gain[len(gain)-15:len(gain)-1]))
```

```python
avg_gain
```

```python
reverse_gain = gain[::-1]
```

```python
for i in range(15,len(gain)):
  avg_gain_values = (13*avg_gain[i-1]+reverse_gain[i-1])/14
  avg_gain.append(avg_gain_values)

len(avg_gain)
```

```python
final_avg_gain = avg_gain[::-1]
new_df['avg_gain'] = final_avg_gain
```

```python
new_df
```

```python
avg_loss = [0 for i in range(14)]
```

```python
avg_loss.append(sum(loss[len(loss)-15:len(loss)-1]))
```

```python
reverse_loss = loss[::-1]
```

```python
for i in range(15,len(loss)):
  avg_loss_values = (13*avg_loss[i-1]+reverse_loss[i-1])/14
  avg_loss.append(avg_loss_values)

len(avg_loss)
```

```python
final_avg_loss = avg_loss[::-1]
new_df['avg_loss'] = final_avg_loss
```

```python
new_df
```

```python
rs = new_df['avg_gain']/new_df['avg_loss']
```

```python
new_df['rs'] = rs
```

```python
new_df
```

```python
new_df['rs'] = new_df['rs'].fillna(0)
```

```python
new_df
```

```python
rsi = 100 - (100/(1+new_df['rs'])) 
```

```python
new_df['rsi'] = rsi
new_df
```

```python
new_df.to_csv('first_pair_with_rsi.csv')
```
