```python
correlation  = new_df.corr()
correlation.style.background_gradient(cmap = 'BrBG')
```

```python
correlation.to_csv("file.csv")
```

```python
corr_pairs = correlation.unstack()
print(corr_pairs)
```

```python
sorted_pairs = corr_pairs.sort_values(kind="quicksort")
print(sorted_pairs)
```

```python
strong_pairs = sorted_pairs[sorted_pairs > 0.85 ]
print(strong_pairs)
```

```python
strong_pairs_final = strong_pairs[strong_pairs < 1.00 ]
print(strong_pairs_final)
```

```python
strong_pairs_final.to_csv("pairs.csv")
```
